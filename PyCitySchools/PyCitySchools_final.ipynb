```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

# District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = len(school_data)

#Calculate the total number of students
total_students = len(student_data)

#Calculate the total budget
total_budget = school_data["budget"].sum()

#Calculate the average math score
average_math_score = student_data["math_score"].mean()

#Calculate the average reading score
average_reading_score = student_data["reading_score"].mean()

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
student_data["average_score"] = (student_data["reading_score"] + student_data["math_score"])/2
overall_pass_rate = student_data["average_score"].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
#passing_math = student_data.loc[student_data["math_score"]>=70]
passing_math = student_data[student_data["math_score"]>=70]
percent_passing_math = (len(passing_math)/total_students)*100

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = student_data[student_data["reading_score"]>=70]
percent_passing_reading = (len(passing_reading)/total_students)*100
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting
Summary_Frame = pd.DataFrame([{"Total Schools":f'{total_schools}',
                             "Total Students":f'{total_students:,}',
                             "Total Budget":f'${total_budget:,.2f}',
                             "Average Math Score":f'{average_math_score:.2f}', 
                             "Average Reading Score":f'{average_reading_score:.2f}',
                             "Overall Passing Rate":f'{overall_pass_rate:.2f}%',
                             "% Passing Math":f'{percent_passing_math:.2f}%',
                             "% Passing Reading":f'{percent_passing_reading:.2f}%'}])
                              
Summary_Frame
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#School Name
#Note:  First Cell must be re-executed inorder for set_index to work properly

#Create an empty DataFrame to store results.  Each Series is added as they are created
#GroupBy object is created to construct calculated series for Averages
school_summary_df = pd.DataFrame(school_data[["school_name","type"]],index=school_data["School ID"])
groupby_school = school_data_complete.groupby("School ID")

#Set Index of school_data to School ID

#Total Students
school_summary_df["Number of Students"] = groupby_school["Student ID"].count()

#Total School Budget
school_summary_df["School Budget"] = school_data["budget"]

#Per Student Budget
school_summary_df["Per Student Budget"] = school_summary_df["School Budget"]/school_summary_df["Number of Students"]

#Average Math Score
school_summary_df["Average Math Score"] = groupby_school["math_score"].mean()

#Average Reading Score
school_summary_df["Average Reading Score"] = groupby_school["reading_score"].mean()

#% Passing Math - I use the apply function and pass a function to filter the groupbby object
#groupby_school.apply(lambda df: len(df))
#groupby_school.apply(lambda df: len(df[df['math_score'] >= 70]))
school_summary_df["% Passing Math"]  = groupby_school.apply(lambda df: (len(df[df['math_score'] >= 70])/len(df))*100)

#% Passing Reading
school_summary_df["% Passing Reading"] = groupby_school.apply(lambda df: (len(df[df['reading_score'] >= 70])/len(df))*100)

#Overall Passing Rate (Average of the above two - use the series that were already created
school_summary_df["Overall Pass Rate"] = (school_summary_df["Average Math Score"]+ school_summary_df["Average Reading Score"])/2

school_summary_df.rename(columns={"type":"School Type"},inplace = True)
```

```python
school_summary_print = school_summary_df.reset_index().set_index("school_name")[["School Type","Number of Students",\
                                                                                 "School Budget","Per Student Budget",\
                                                                                 "Average Math Score","Average Reading Score",\
                                                                                 "% Passing Math","% Passing Reading",\
                                                                                 "Overall Pass Rate"]]
#school_summary_print.drop("School ID", axis=1, inplace = True)
school_summary_print.index.name = None
school_summary_print
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
#spending_bins = [0, 585, 615, 645, 675]
#group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
mean_summary = school_summary_df["Per Student Budget"].mean()
std_summary = school_summary_df["Per Student Budget"].std()
bin_list = [int(mean_summary-(2-std_dev)*std_summary) for std_dev in range(5)]
my_summary_bins = bin_list
my_group_names = [f"< ${bin_list[1]}", f"\${bin_list[1]}-\${bin_list[2]}",f"\${bin_list[2]}-\${bin_list[3]}", f"> \${bin_list[3]}"]
```

```python
school_summary_print["Budget Category"]  = pd.cut(school_summary_print["Per Student Budget"],my_summary_bins, labels = my_group_names, include_lowest = True)
by_spending = school_summary_print.groupby("Budget Category").mean()[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Pass Rate"]]
by_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
#size_bins = [0, 1000, 2000, 5000]
#group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
mean_summary = school_summary_df["Number of Students"].mean()
std_summary = school_summary_df["Number of Students"].std()
bin_list = [int(mean_summary-(2-std_dev)*std_summary) for std_dev in range(5)]
my_summary_bins = bin_list
my_group_names = [f"< {bin_list[1]} students", f"{bin_list[1]}-{bin_list[2]} students",f"{bin_list[2]}-{bin_list[3]} students", f"> {bin_list[3]} students"]
```

```python
school_summary_print["Size Category"] = pd.cut(school_summary_print["Number of Students"],my_summary_bins, labels = my_group_names, include_lowest = True)
by_size = school_summary_print.groupby("Size Category").mean()[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Pass Rate"]]
by_size
```
